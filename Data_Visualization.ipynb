```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
```

## Line Plot

```python
a = np.arange(1, 10)

plt.plot(a)
plt.show()
```

```python
a = np.arange(1, 10)
b = a+5

plt.plot(a, b)
plt.show()
```

### 라인색상 변경하기

```python
a = np.arange(1, 10)
b = a+5

plt.plot(a, b, 'r') # 라인색상 'r' 부여
plt.show()
```

### 라인의 종류 변경하기

```python
a = np.arange(1,10)
b = a+5

plt.plot(a,b,'or')
plt.show()
```

### 축 이름 지정하기

```python
a = np.arange(1, 10)
b = a+5

plt.plot(a, b)
plt.xlabel('a Label')
plt.ylabel('b Label')
plt.title('a, b Label Test')

plt.show()
```

### 그래프 사이즈 지정하기
