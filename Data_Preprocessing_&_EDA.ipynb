```python
url = ('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/kt%26g/kt%26g_0.csv')
df = pd.read_csv(url)
df
```

```python
df = df[['19/6','12,578','3,991']]
df
```

```python
df.columns = ['년/월', '합격자 수', '불합격자 수']
df
```

```python
df.dtypes
```

```python
# str(object)을 int로 바꾸는 함수
def toint(string) :
    return int(string.replace(',',''))

df['합격자 수'] = df['합격자 수'].apply(toint)
df['불합격자 수'] = df['불합격자 수'].apply(toint)
df.dtypes
```

```python
df['년/월'] = df['년/월'].apply(toint)
```
